```python
nombre = input("Ingrese su nombre: ")
PrimerApellido = input("Ingrese su primer apellido: ")
SegundoApellido = input("Ingrese su segundo apellido: ")
DiaNacimiento = input("Ingrese su día de nacimiento: ")
MesNacimiento = input("Ingrese su mes de nacimiento: ")
YearNacimiento = input("Ingrese su año de nacimiento: ")
Sexo = input("Ingrese su sexo: ")
estado = input("Ingrese su estado: ")

letraNombre = nombre[0]
letraPrimerAP = PrimerApellido[0:1]
letraSegundoAP = SegundoApellido[0]
numYearNacimiento = YearNacimiento[2:4]
numMesNacimiento = MesNacimiento.zfill(2)
numDiaNacimiento = DiaNacimiento.zfill(2)
letraSexo = Sexo[0]
letraEstado = estado[:2]

curp = letraPrimerAP + letraSegundoAP + letraNombre + numYearNacimiento + numMesNacimiento + numDiaNacimiento + letraSexo + letraEstado

print("Su CURP es:", curp.upper())
```
