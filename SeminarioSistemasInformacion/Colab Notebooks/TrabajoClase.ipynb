```python
while True:

    entrada = input("Ingrese una cadena de caracteres o 'fin' para salir: ")

    # Verifica si el usuario desea salir del programa
    if entrada == 'fin':
        break

    try:
        #leer palabras con la libreria split que
        num_palabras = len(entrada.split())
        #leer vocales
        num_vocales = sum(1 for letra in entrada if letra in "a,e,i,o,u,A,E,I,O,U")

        print("vocales",num_vocales)
        print("plabras",num_palabras)
    except ValueError:
        # Maneja el error si la entrada no es un número
        print("Error: Ingrese un una opcion válido.")

```
