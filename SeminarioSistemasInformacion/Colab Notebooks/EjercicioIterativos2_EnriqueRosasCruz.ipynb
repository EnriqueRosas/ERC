```python
total = 0
cantidad_numeros = 0

while True:

    entrada = input("Ingrese un número o escriba 'fin' para salir: ")

    # Verifica si el usuario desea salir del programa
    if entrada.lower() == 'fin':
        break

    try:
        # Intenta convertir la entrada a un número
        numero = float(entrada)
        # Actualiza el total y la cantidad de números
        total += numero
        cantidad_numeros += 1
    except ValueError:
        # Maneja el error si la entrada no es un número
        print("Error: Ingrese un número válido.")

# Se verifica si se ingresaron números antes de mostrar resultados
if cantidad_numeros > 0:
    # Calcula la media
    media = total / cantidad_numeros

    print(f"\nTotal: {total}")
    print(f"Cantidad de números: {cantidad_numeros}")
    print(f"Media: {media}")
else:
    print("No se ingresaron números.")
```
