```python
numero = int(input("Ingrese un número entero de 3 cifras: "))
if not (0 <= numero <= 999) and not (-999 <= numero <= 0):
    print("Error: Debe ingresar un número entero de 3 cifras.")
else:
  digitos = [int(digito) for digito in str(abs(numero))]
  digitos.sort(reverse=True)
  numero_formado = int(''.join(map(str, digitos)))
  if numero < 0:
      numero_formado *= -1
  print("La cifra mayor es:", numero_formado)
```
