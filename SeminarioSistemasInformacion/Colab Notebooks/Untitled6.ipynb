```python

#suma, resta, traspuesta
#funcimes- visualizar, ingresar
# Definir dos matrices
matriz1 = []
matriz2 = []
matriz3 = []
Rows = input("ingrese la cantidad de filas")
Colws= input("ingrese la cantidad de columnas")


# Ingresar la primera matriz
print("Ingrese la primera matriz:")
for i in range(3):  # Supongamos matrices 3x3
    fila = [int(input()) for _ in range(3)]
    matriz1.append(fila)

# Ingresar la segunda matriz
print("Ingrese la segunda matriz:")
for i in range(3):  # Supongamos matrices 3x3
    fila = [int(input()) for _ in range(3)]
    matriz2.append(fila)


# Inicializar una matriz para almacenar la suma
# la matriz debe ser del mismo tama√±o
suma_matriz = [[0,0,0],[0,0,0],[0,0,0]]

# Sumar las matrices

for i in range(len(matriz1)):
    for j in range(len(matriz1[0])):
        suma_matriz[i][j] = matriz1[i][j] + matriz2[i][j]

# Imprimir la matriz resultante
print("el resultado de la suma de las matrices anteriores es :")
print(suma_matriz)
```
