```python
#ejercicio en clase de listas
#elaborar un programa que permita
#ingresar 10 números que serán almacenados
#después imprimirlos en pantalla
#ademas calcular las medias de tendencia
#central media

numeros = []

while True:
    # Ingresar un número
    numero = input("Ingrese un número o escriba 'fin' para terminar: ")

    # Verificar si el usuario quiere salir del bucle
    if numero == 'fin':
        break

    # Convertir a número flotante y agregar a la lista
    try:
        numero = float(numero)
        numeros.append(numero)
    except ValueError:
        print("Por favor, ingrese un número válido.")

# Imprimir lista
print("\nNúmeros ingresados:")
print(numeros)

# Calcular media aritmética
if len(numeros) > 0:
    media_aritmetica = sum(numeros) / len(numeros)
    print(f"\nMedia Aritmética: {media_aritmetica}")
else:
    print("No se ingresaron números para calcular la media aritmética.")

# Calcular la mediana
numeros.sort()
longitud = len(numeros)
mediana = 0

if longitud > 0:
    if longitud % 2 == 0:
        # Si la cantidad de números es par, la mediana es el promedio de los dos números centrales
        mediana = (numeros[longitud // 2 - 1] + numeros[longitud // 2]) / 2
    else:
        # Si la cantidad de números es impar, la mediana es el número central
        mediana = numeros[longitud // 2]

    print(f"Mediana: {mediana}")
else:
    print("No se ingresaron números para calcular la mediana.")

```
