```python
while True:
    try:
        numero_ingresado = int(input("Ingrese un número entero de 3 cifras (o ingrese 0 para salir): "))
        if numero_ingresado == 0:
            print("Saliendo del programa.")
            break
        elif not (100 <= abs(numero_ingresado) <= 999):
            print("Error: Debe ingresar un número entero de 3 cifras.")
        else:
              digitos = [int(digito) for digito in str(abs(numero_ingresado))]
              digitos.sort(reverse=True)
              numero_formado = int(''.join(map(str, digitos)))
              if numero_ingresado < 0:
                  numero_formado *= -1
              print("La cifra mayor es:", numero_formado)

    except ValueError:
        print("Error: Ingrese un número entero válido.")
```
